## Minimum detectable line flux

This script calculates the minimum detectable line flux (MDLF) of DESHIMA 2.0.

- The user can change some parameters related to observation.
- After execution, the plot will be updated, and the sensitivity curves for baseline and goal performances can be downloaded as a csv file.
- The calculation is based on the DESHIMA 2.0 hardware requirements, which can be found [here](https://deshima.kibe.la/shared/entries/ff96eb1f-ef80-41e2-adf8-178769a72e3f).

```python
MDLF_simple(
    F = np.logspace(np.log10(220),np.log10(440),349)*1e9, # R = 500 means 349 channels in 220-440 GHz
    pwv = 0.5,      # Precipitable Water Vapor in mm (0.1 < pwv < 2.0 mm)
    EL = 60.,       # Elevation angle in degrees
    snr = 5.,       # Target S/N of the detection
    obs_hours = 8.  # Observation time in hours, including ON-OFF and calibration overhead
)
```
